```python
# Импортиране на нужните библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
TICKERS = ['MSFT', 'AAPL', 'GOOGL', 'AMZN', 'META']
```

```python
all_data = []
for ticker in TICKERS:
    df = yf.download(ticker, start='2020-01-01')
    df['Ticker'] = ticker
    all_data.append(df)

stocks_df = pd.concat(all_data)
stocks_df.reset_index(inplace=True)
print(stocks_df.head())

```

```python
# Преобразуване на MultiIndex DataFrame в long format
stocks_df = stocks_df.stack(level=1).reset_index()
stocks_df = stocks_df.rename(columns={"level_1": "Ticker"})
print(stocks_df.head())
```

```python
# Ако вече има колона 'Ticker', премахваме я преди преобразуването
if 'Ticker' in stocks_df.columns:
    stocks_df = stocks_df.drop(columns=['Ticker'])

stocks_df = stocks_df.stack(level=1).reset_index()
stocks_df = stocks_df.rename(columns={"level_1": "Ticker"})
print(stocks_df.head())
```
